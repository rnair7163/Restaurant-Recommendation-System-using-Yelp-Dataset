```python
import pandas as pd
import json,os
import numpy as np
```

```python
users = []
path = 'D://DPA Data//user.json'
with open(os.path.join(os.getcwd(), path),encoding="utf8") as f:
    for line in f:
        users.append(json.loads(line))
print("Read {count} users".format(count=len(users)))
```

```python
users = pd.DataFrame(users)
users.head()
```

```python
reviews = pd.read_csv('reviews.csv')
reviews.head()
```

```python
user_ids = reviews.user_id.unique()
user_ids = np.array(user_ids)
len(user_ids)
```

```python
users = users[users['user_id'].isin(user_ids)]
users.shape
```

```python
users.describe()
```

```python
# fixing indexes
users.index = range(users.shape[0])
users.head()
```

```python
## Checking null values
users.isnull().sum()
```

```python
print(users.yelping_since.head())
len(users.yelping_since.unique())

## Too many unique values because of the timestamp
```

```python
yelping_since = [x.split(' ')[0] for x in users.yelping_since]
len(np.unique(yelping_since))
```

```python
users.yelping_since = yelping_since
len(users.yelping_since.unique())
```
